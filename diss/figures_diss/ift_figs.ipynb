```python
import numpy as np
import matplotlib.pyplot as plt
import tikzplotlib
```

```python
y1 = lambda x: np.sqrt(1-x**2)
y2 = lambda x: -np.sqrt(1-x**2)
```

```python
x = np.linspace(-1, 1, 1000)
```

```python
f = plt.figure()
plt.plot(x, y1(x))
plt.plot(x, y2(x))
tikzplotlib.save('unitcircle.pgf')
```
