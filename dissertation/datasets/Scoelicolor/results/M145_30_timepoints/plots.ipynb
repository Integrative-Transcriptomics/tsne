```python
import sys
wd = '/ceph/ibmi/it/users/zabel/tsne/diss'
sys.path.insert(0, wd)

import numpy as np
import pandas as pd

from utils import plot_heatmap
import matplotlib.pyplot as plt
import tikzplotlib

from tueplots import cycler, fonts, fontsizes, bundles
from tueplots.constants import markers
from tueplots.constants.color import palettes

plt.rcParams['figure.dpi'] = 600
plt.rcParams.update(cycler.cycler(color=palettes.tue_plot))
```

```python
Y_star = np.load('/ceph/ibmi/it/users/zabel/tsne/diss/datasets/Scoelicolor/results/M145_30_timepoints/tsne_embedding.npy')
X = pd.read_csv('/ceph/ibmi/it/users/zabel/tsne/diss/datasets/Scoelicolor/results/M145_30_timepoints/f199_full_genes.tsv', header=0, sep='\t', index_col=0)
labels_str = X.columns
labels_int = [int(i) for i, j in enumerate(labels_str)]
labels_unique = ['t'+i for i in labels_str]
labels = labels_unique
```

```python
import matplotlib
cmap = matplotlib.colormaps.get_cmap('rainbow')

# Pick N numbers from 0 to 1
gradient = np.linspace(0, 1, len(labels_int))
colors = [matplotlib.colors.to_hex(cmap(i)) for i in gradient]

f = plt.figure()
for i, j in enumerate(Y_star):
  plt.plot(j[0], j[1], c = colors[i], marker='o')
plt.plot(Y_star[:, 0], Y_star[:, 1], c='lightgray')
for i, txt in enumerate(labels):
  plt.annotate(txt, (Y_star[i, 0], Y_star[i, 1]))
plt.xlabel('TSNE 1')
plt.ylabel('TSNE 2')
#plt.legend(handles=scatter.legend_elements(num=8)[0], labels=labels)
#plt.legend()
tikzplotlib.save("/ceph/ibmi/it/users/zabel/tsne/diss/datasets/Scoelicolor/results/M145_30_timepoints/tsne.pgf")
```
