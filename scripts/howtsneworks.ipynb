```python
import gzip
import pickle

import sys, os
from os import path
import numpy as np
import seaborn as sns
import scipy.stats as stats
from sklearn import datasets


import matplotlib.pyplot as plt
from tueplots import figsizes, fonts, bundles
from tueplots import cycler
from tueplots.constants import markers
from tueplots.constants.color import palettes

# Increase the resolution of all the plots below
plt.rcParams.update({"figure.dpi": 150})
```

```python
mu = 0
variance = 1
sigma = np.sqrt(variance)
t = np.linspace(mu - 5*sigma, mu + 5*sigma, 100)

with plt.rc_context(bundles.beamer_moml()):
    plt.rcParams.update({'figure.figsize': (0.6*5.511811263318113, 1*2.480315068493151)})
    f, ax1 = plt.subplots()
    ax1.plot(t, stats.norm.pdf(t, mu, sigma), label='Gaussian')
    ax1.plot(t, stats.t.pdf(t, df=1), label='StudentÂ´s t')
    plt.legend()
    plt.savefig('gaussandstudent.pdf')
```

```python
x, y = datasets.make_blobs(n_samples=6, n_features=2, centers=((-2, 2), (0, -1)), cluster_std=1)
midpoint = (-1.7, 1)
d = dist = [np.linalg.norm(i-midpoint) for i in x]
with plt.rc_context(bundles.beamer_moml()):
    plt.rcParams.update({'figure.figsize': (0.6*5.511811263318113, 1*2.480315068493151)})
    f, (ax1, ax2) = plt.subplots(2, 1) 
    ax1.scatter(x[:,0], x[:, 1], label='js')
    ax1.scatter(midpoint[0], midpoint[1], label='i')
    for i in range(0, len(x)):
        ax1.plot((midpoint[0], x[i, 0]), (midpoint[1], x[i, 1]), '-', c=palettes.tue_plot[2])
    ax1.legend()
    ax2.scatter(0, 0, color=palettes.tue_plot[1], label='i')
    ax2.scatter(d, np.repeat(0, len(d)), label='js')
    ax2.plot(t, stats.norm.pdf(t, mu, sigma), label='Gaussian', c=palettes.tue_plot[2])
    ax2.legend()
    plt.savefig('xtop.pdf')
```
