# Jacobian-Vector-product

```python
from jax import vjp, jvp
from functools import partial
```

```python
# first version
def mixed_Jacobian_vector_product(f, primals, tangents):
    X, Y = primals
    first = lambda X: jacfwd(f, argnums=1)(X, Y)
    second, second_t = jvp(first, (X,), (np.eye(len(X))[0],)) # --> first column of Jacobian!!!
    return second, second_t
```

```python
# second version (probably faster due to less loops when traversing through np.eye when constructing full jacobian)
# but cov_X is in dimension that first version is better?
def mixed_Jacobian_vector_product(f, primals, tangents):
    X, Y = primals
    first = lambda Y: jacfwd(f, argnums=0)(X, Y)
    second, second_t = jvp(first, (Y,), (np.eye(len(Y))[0],)) # --> first row of Jacobian!!!
    return second, second_t
```

# Neumann approximation of Hessian

```python
def approxInverseHVP(f, primals, v):
    p = v
    print(v)
    X, Y = primals
    first = lambda Y: jacfwd(f, argnums=1)(X, Y)
    for i in range(20):
        dif = jvp(first, (Y,), (v,))[1]
        #print(dif)
        v = v - dif
        p = p - v
    return p
```

```python
def HVP(f, primals, v):
    X, Y = primals
    first = lambda Y: jacfwd(f, argnums=1)(X, Y)
    return jvp(first, (Y,), (v,))[1]
```

```python
def approx_inv(M, iteration):
    # https://jwcn-eurasipjournals.springeropen.com/articles/10.1186/s13638-019-1631-x#Sec10
    I = np.eye(len(M))
    D = np.linalg.inv(np.diag(np.diag(M)))
    Z = D @ (3 * I - M @ D @ (3 * I - M @ D))
    s = np.zeros(M.shape)
    for i in range(1, iteration):
        s += (I - Z @ M)**i
    return s @ Z
H_inv = approx_inv(H, 10)
f = plt.figure()
plt.imshow(H_inv)
plt.colorbar()
f = plt.figure()
plt.imshow(np.linalg.inv(H))
plt.colorbar()
```
